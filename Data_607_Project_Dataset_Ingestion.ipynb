```python
import pandas as pd

image_csv = BASE_PATH + "FEIS_Friends_of_Elk_Island_2020_image_report.csv"
df_image = pd.read_csv(image_csv)

print(df_image.shape)
print(df_image.columns.tolist())
df_image.head()
```

```python
tag_csv = BASE_PATH + "FEIS_Friends_of_Elk_Island_2020_tag_report.csv"
df_tag = pd.read_csv(tag_csv)

print(df_tag.shape)
print(df_tag.columns.tolist())
df_tag.head()
```

```python
df_image_small = df_image[[
    "image_id",
    "location",
    "location_id",
    "image_date_time",
    "media_url"
]]

df_tag_small = df_tag[[
    "image_id",
    "species_common_name",
    "age_class",
    "sex_class"
]]
```

```python
df_merged = pd.merge(df_image_small, df_tag_small, on="image_id", how="inner")
```

```python
print(df_merged.shape)
df_merged.head()
```

```python
df_merged["species_common_name"].value_counts().head(20)
```

##Data Cleaning

```python
df_clean = df_merged.copy()

#Removing bad labels
df_clean = df_clean.dropna(subset=["species_common_name"])
df_clean = df_clean[df_clean["species_common_name"] != "NONE"]
df_clean = df_clean[df_clean["species_common_name"] != "STAFF/SETUP"]
df_clean = df_clean[df_clean["species_common_name"] != "Unidentified"]

#Removing bad URLS(but there weren't any for this dataset)
df_clean = df_clean.dropna(subset=["media_url"])

print(df_clean.shape)
df_clean["species_common_name"].value_counts().head(20)
```

```python
#Selecting species
selected_species = [
    "Elk (wapiti)",
    "Bison",
    "White-tailed Deer",
    "Moose",
    "Coyote",
    "Snowshoe Hare"
]

df_project = df_clean[df_clean["species_common_name"].isin(selected_species)].copy()

print(df_project.shape)
print(df_project["species_common_name"].value_counts())
```

##Downloading Images

```python
#Picked 300 per sample for initial testing purpose
n_per_class = 300

df_sampled = (
    df_project
    .groupby("species_common_name", group_keys=True)
    .apply(lambda x: x.sample(n=n_per_class, random_state=42))
    .reset_index(drop=False)
    .drop("level_1", axis=1)
)

print(df_sampled.shape)
print(df_sampled.head())
print(df_sampled["species_common_name"].value_counts())
```

```python
df_sampled["age_class"].value_counts()
```

```python
df_sampled["sex_class"].value_counts()
```

```python
#Making filepath so that dataframe becomes the source of truth(label)
all_images_path = BASE_PATH + "all_images"

df_sampled["filepath"] = df_sampled["image_id"].astype(str).apply(
    lambda x: all_images_path + x + ".jpg"
)
```

```python
df_sampled.to_csv(BASE_PATH + "dataset_labels.csv", index=False)
```

```python
import os
os.makedirs(all_images_path, exist_ok=True)
print(all_images_path)
```

```python
print(df_sampled.head())
```

```python
import requests

success = 0
failed = 0

for i, row in df_sampled.iterrows():
    url = row["media_url"]
    image_id = row["image_id"]

    save_path = all_images_path + "/" + str(image_id) + ".jpg"
    
    try:
        r = requests.get(url)

        if r.status_code == 200:
            with open(save_path, "wb") as f:
                print(i)
                f.write(r.content)
                print(save_path)
            success += 1
        else:
            print("Failed:", image_id)
            failed += 1

    except:
        print("Error:", image_id)
        failed += 1

print("Downloaded:", success)
print("Failed:", failed)
```
